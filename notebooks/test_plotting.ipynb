```python
import pickle
from lgsm.plotting import plot_sed
import matplotlib.pyplot as plt
import numpy as np
```

```python
with open("simulated_seds.pkl", "rb") as file:
    seds = pickle.load(file)
```

```python
fig, ax = plot_sed(seds["wave"], seds["sed_mag"], sed_unit="mag")
```

```python
fig, axes = plt.subplots(
    1, 2,
    figsize=(7, 3.5), 
    dpi=120, 
    constrained_layout=True
)

settings = {
    "wave": seds["wave"],
    "sed": seds["sed_mag"][[0, 400]],
    "z": np.arange(3),
    "normalize_at": None,
    "sed_unit": "mag",
    "ax_settings": {
        "xscale": "log",
    },
    "plot_settings": {
        "c": np.repeat(["C0", "C1"], 3),
        "ls": ["-", "--", ":"],
    },
}

# plot the SEDs in AB magnitude
plot_sed(ax=axes[0], plot_unit="mag", **settings)
axes[0].invert_yaxis()

# plot the SEDs in F_lambda
plot_sed(ax=axes[1], plot_unit="flambda", **settings)

plt.show()
```
